```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=400,noise=0.02)
```

```python
plt.scatter(X[:,0],X[:,1], c = y)
plt.show()
```

```python
def phi(X):
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X1**2 + X2**2
    
    Xn = np.zeros((X.shape[0],3))
    Xn[:,:-1] = X
    Xn[:,-1] = X3
    
    return Xn
```

```python
X_ = phi(X)
```

## Logistic Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression()
```

```python
cross_val_score(lr,X,y, scoring="accuracy",cv=5).mean()*100
```

###  Custom Kernel in SVM

```python
def custon_kernel(x1,x2):
    return np.square(np.dot(x1,x2.T))
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel=custon_kernel)
```

```python
svc.fit(X,y)
```

```python
cross_val_score(svc,X,y,scoring="accuracy",cv=5).mean()
```
